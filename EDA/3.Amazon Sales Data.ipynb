Q1. What is the average rating for each product category ?

```python
df.columns
```

```python
# Convert the 'rating' column to numeric, forcing errors to NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate the average rating for each product category
avg_rating_per_category = df.groupby('category')['rating'].mean().head(10)
print(avg_rating_per_category)

# Plot the average rating for each product category
plt.figure(figsize=(10,8))
sns.barplot(x=avg_rating_per_category.index, y=avg_rating_per_category.values)
plt.title("Average Rating per Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.xticks(rotation=90)
plt.show()
```

```python
# 1. **Highest Average Rating**: The category "Computers&Accessories|Accessories&Peripherals|Cables&Accessories|Cables|DVICables" and "Computers&Accessories|Accessories&Peripherals|Cables&Accessories|Cables|EthernetCables" have the highest average rating of 4.4.

# 2. **Lowest Average Rating**: The category "Computers&Accessories|Accessories&Peripherals|Audio&VideoAccessories|PCHeadsets" has the lowest average rating of 3.5.

# 3. **General Trend**: Most categories have an average rating above 4.0, indicating that the products in these categories are generally well-received by customers.

# 4. **Variation**: There is some variation in the average ratings across different categories, suggesting that certain categories might have more consistent product quality or customer satisfaction than others.

# These insights can help in understanding customer satisfaction across different product categories and identifying areas for improvement.
```

Q4. How does the average discount percentage vary across categories?


```python
# Remove the '%' sign and convert 'discount_percentage' to numeric
df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype(float)

# Calculate the average discount percentage for each category
avg_discount_per_category = df.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).head(10)
print(avg_discount_per_category)

# Plot the average discount percentage for each category
plt.figure(figsize=(10,8))
sns.barplot(x=avg_discount_per_category.values, y=avg_discount_per_category.index, palette='viridis')
plt.title("Average Discount Percentage per Category")
plt.xlabel("Average Discount Percentage")
plt.ylabel("Category")
plt.show()
```

```python
# Insight >>
# Categories like "Phone Charms" and "Cable Connection Protectors" have the highest average discount percentages, 
# indicating significant price reductions in these categories.
```

Q6. 6. What are the most popular product keywords?

```python
from collections import Counter
import re

# Extract keywords from product names
keywords = df['product_name'].str.lower().str.cat(sep=' ')
keywords = re.findall(r'\b\w+\b', keywords)

# Count the frequency of each keyword
keyword_counts = Counter(keywords)

# Get the most common keywords
most_common_keywords = keyword_counts.most_common(10)
print(most_common_keywords)

# Convert to DataFrame for better visualization
keywords_df = pd.DataFrame(most_common_keywords, columns=['Keyword', 'Count'])

# Plot the most common keywords
plt.figure(figsize=(10,8))
sns.barplot(x='Count', y='Keyword', data=keywords_df, palette='coolwarm')
plt.title("Top 10 Most Popular Product Keywords")
plt.xlabel("Count")
plt.ylabel("Keyword")
plt.show()
```

Q8. What is the correlation between discounted_price and rating?

```python
# Remove the currency symbol and convert to numeric
df["discounted_price"] = df["discounted_price"].replace('[₹,]', '', regex=True).astype(float)

correlation = df["discounted_price"].corr(df["rating"])
print(f"Correlation between Discounted Price and Rating: {correlation:.2f}")

sns.heatmap(df[["discounted_price", "rating"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
